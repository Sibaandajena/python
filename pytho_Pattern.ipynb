## Pattern print using conditional statement and loops

### 1.Right Angle Triangle Pattern

```python
for i in range(1,6):
    print(i*'* ')
    i + 1
```

### 2. Inverted Right Angle Triangle Pattern

```python
for i in range(5,0,-1):
    print(i*'* ')
```

### 3. Pyramid pattern

```python
for i in range(1,6):
    print(''*(5-i)+ ' * ' *(2*i-1) )
```

### 4.Inverted Pyramid

```python
for i in range(5,0,-1):
    print(''*(5-i)+ ' * ' *(2*i-1) )
```

### 5. Diamond Pattern

```python
for i in range(1,6):
    print(''*(5-i)+ ' * ' *(2*i-1) )

for i in range(4,0,-1):
    print(''*(5-i)+ ' * ' *(2*i-1) )
```

### 6. Hallow Square Pattern

```python
for i in range (5):
    for j in range(5):
        if i==0 or i==4 or j==0 or j==4:
            print('*', end='')
        else:
            print(' ', end='')
            
    print()
```

### 7. Full Square Pattern

```python
for i in range(5):
    print(' * ' * 5)
```

### 8. Right Angle Triangle (Number print)

```python
for i in range(1,6):
    print(i* '* ') 
    i + 1
```
